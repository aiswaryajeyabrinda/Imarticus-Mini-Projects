# Cleaning the data

```python
data.isnull().sum()
```

```python
data.normloss.fillna(data.normloss.mean(),inplace=True)
```

```python
data.stroke.fillna(data.stroke.mean(),inplace=True)
```

```python
data.horsepower.fillna(data.horsepower.mean(),inplace=True)
```

```python
data.bore.fillna(data.bore.median(),inplace = True)
```

```python
data.peakrpm.fillna(data.peakrpm.mean(),inplace = True)
```

```python
data.price.fillna(data.price.median(),inplace = True)
```

```python
data.numofdoor.fillna(data.numofdoor.mode,inplace=True)
```

```python
q3=data.normloss.quantile(0.75)
q1=data.normloss.quantile(0.25)
iqr=q3-q1
```

```python
UE=q3+1.5*(iqr)
LE=q1-1.5*(iqr)
```

```python
data[(data.normloss<UE)&(data.normloss>LE)]
```

```python
data.normloss[data.normloss>UE]=UE
data.normloss[data.normloss<LE]=LE
```
