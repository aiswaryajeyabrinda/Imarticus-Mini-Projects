# Checking for linearity

```python
for i in data.columns.drop(['medv']):
        print("medv vs",i)
        sns.scatterplot(data=data,x=data['medv'],y=data[i])
        plt.show()
```

```python
data.corr()
```

To check if the correlation is significant or not, We do Pearsonr
</br>

Null Hypothesis - H0 : There is No significant linear relationship(correlation) between x and y in the population
</br>

Alternate Hypothesis - H1 : There is a significant linear relationship(correlation) between x and y in the population

```python
from scipy.stats import pearsonr
```

```python
for i in data.columns.drop('medv'):
    print("\nCorrelation between medv and",i)
    correlation,p_value=0,0
    correlation,p_value=pearsonr(data[i],y)
    print("Correlation = ",correlation,"\np_value = ",p_value)
    if p_value>0.05:
        print("Not Significant.")
    else:
        print("Significant.")
```

### Checking for Normality of Dependent Variable(y) 

```python
sns.kdeplot(y)
plt.show()
```

Dependent variable is not normally distributed
</br>

To check the normality of the dependent variable, we do Shapiro-Wilk test

```python
from scipy import stats
```

```python
statistic,p = stats.shapiro(y)
print("Statistic = ",statistic,"\nP_value = ",p)
if p>0.05:
    print("Significant.")
else:
    print("Not Significant.")
```
