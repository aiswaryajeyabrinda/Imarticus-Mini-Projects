```python
from tensorflow.keras.preprocessing import sequence
```

```python
train_words = sequence.pad_sequences(train_words,maxlen = 500)
```

```python
test_words = sequence.pad_sequences(test_words,maxlen=500)
```

```python
from keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN
from tensorflow.keras import layers
from keras.layers import Embedding,Activation
model = Sequential()
```

```python
model.add(layers.Embedding(input_dim = 10000, output_dim = 32))
model.add(SimpleRNN(units=32,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(loss= 'binary_crossentropy',optimizer = 'adam', metrics=['accuracy'])
```

```python
model.fit(train_words,train_labels)
```

```python
model.evaluate(test_words,test_labels)
```
